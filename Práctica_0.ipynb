- La estructura de paquetes en ROS 2.

ROS (Robot Operating System) es un middleware de código abierto que proporciona servicios estándar para el desarrollo de aplicaciones robóticas, como la comunicación entre procesos, el control de hardware y herramientas de desarrollo.

A diferencia de ROS 1, ROS 2 fue diseñado para mejorar la confiabilidad, seguridad y compatibilidad con sistemas embebidos, lo que permite su uso en aplicaciones industriales.

1. El Espacio de Trabajo (Workspace)
El espacio de trabajo es el directorio raíz donde organizamos nuestros proyectos de robótica. A diferencia de versiones anteriores, en ROS 2 la estructura estándar tras compilar con colcon es:

src/: El espacio de código fuente. Aquí es donde clonamos los repositorios y creamos nuestros propios paquetes. Es el directorio donde se desarrolla la mayor parte del código del proyecto.

install/: Aquí es donde ROS 2 coloca los ejecutables y librerías una vez compilados. Es la carpeta que "activamos" (source) para que la terminal reconozca nuestros nodos.

build/: Almacena archivos intermedios generados durante la compilación.

log/: Contiene registros de depuración sobre el proceso de compilación.

2. Anatomía de un Paquete en ROS 2 (Python)
Los paquetes son la unidad fundamental de organización. En nuestra práctica, un paquete de Python tiene los siguientes componentes críticos:

package.xml: El "DNI" del paquete. Contiene metadatos (nombre, versión, autor) y las dependencias (otros paquetes que nuestro código necesita para correr).

setup.py: Sustituye al antiguo CMakeLists.txt para paquetes de Python. Define cómo se instala el código y los entry points (puntos de entrada) para ejecutar los nodos desde la terminal.

resource/: Contiene un archivo marcador para que las herramientas de ROS localicen el paquete.

test/: Directorio para pruebas de integración y funcionalidad.

Subcarpeta del paquete: Donde residen los archivos .py (como nuestros nodos Publisher y Subscriber).



- Definición de Nodos de Tópicos.


1. El Grafo de ROS (ROS Graph)
El sistema robótico no es un programa lineal, sino una red de elementos que procesan datos simultáneamente. Esta red se denomina "Grafo". Engloba a todos los ejecutables y las conexiones (hilos invisibles) que permiten que la información fluya desde los sensores hasta los actuadores.

2. Nodos: La Unidad Funcional
Un Nodo es un proceso modular y único encargado de una tarea específica (ej. leer un sensor LiDAR, controlar la potencia de un motor o procesar una imagen de OpenCV).

Modularidad: Un robot complejo se divide en muchos nodos pequeños para que sea más fácil de reparar y programar.

Ejecución: Un solo archivo de Python puede contener uno o varios nodos.

Gestión del Ciclo de Vida: Los nodos pueden tener estados (Configurado, Activo, Inactivo) para asegurar que el robot no se mueva si un componente falla.

3. Tópicos: El Bus de Datos
Los Tópicos son los canales de comunicación (bus) que conectan a los nodos. Funcionan bajo el modelo de Publicación/Suscripción:

Publisher (Publicador): El nodo que genera y envía los datos (ej. el nodo de la cámara publicando fotogramas).

Subscriber (Suscriptor): El nodo que recibe y utiliza los datos (ej. el nodo de SLAM que recibe los fotogramas).

Mensajería Anónima: Un publicador no necesita saber quién lo escucha; simplemente pone el mensaje en el "bus" (tópico) y cualquier nodo interesado puede tomarlo.

Tipos de Mensaje: Cada tópico tiene un nombre único (ej. /cmd_vel) y un tipo de dato estricto (ej. geometry_msgs/Twist).

4. Herramientas de Inspección y Diagnóstico
Para verificar que el paquete y la estructura funcionan correctamente, utilizamos comandos de terminal:

ros2 node list: Muestra los nodos activos en el grafo.

ros2 topic list: Muestra los canales de comunicación disponibles.

ros2 node info <nombre>**: Muestra con qué tópicos y servicios interactúa un nodo específico.

ros2 topic echo <nombre>**: Permite ver "en vivo" los datos que viajan por una carretera de información.

Nodo Publicador

```python
import rclpy
from rclpy.node import Node
from std_msgs.msg import Float64
from math import sin

class MyNodo(Node):
    def _init_(self):
        super()._init_('my_node')
        self.number = 2
        self.publisher_ = self.create_publisher(
            msg_type = Float64,
            topic = "rpm_topic", 
            qos_profile = 10)
        
        self.amplitude = 0.50
        self.frequency = 1.0
        self.time_counter = 0.0

        self.timer_period = 0.1

        self.timer_ = self.create_timer(
            self.timer_period, callback = self.cbck)
        self.get_logger().info('Nodo de RPM con seno activado')
        
    def cbck(self):
        msg = Float64()

        val=self.amplitude*sin(2*3.1416*self.frequency*self.time_counter)+0.50

        msg.data = val
        self.publisher_.publish(msg)
        self.time_counter += self.timer_period

def main(args= None):
    rclpy.init(args=args)
    node = MyNodo()
    rclpy.spin(node)
    rclpy.shutdown()

if _name== '_main':
    main()
```

```python
import rclpy
from rclpy.node import Node
from std_msgs.msg import Float64

class NodeCounter(Node):
    def _init_(self):
        super()._init_("subscriber_node")
        self.counter_= 0
        self.create_subscription(
            msg_type=Float64, 
            callback= self.sub_cbck, 
            topic= "rpm_topic", 
            qos_profile= 10)
        
        self.publisher_counter_ = self.create_publisher(
            msg_type= Float64, 
            topic='counter_topic',
            qos_profile=10
        )
        self.get_logger().info('Nodo de rad/s activo')
        
    def sub_cbck(self, msg):
        rpm = msg.data

        rads=rpm*(2.0*3.14159/60.0)

        new_msg = Float64()
        new_msg.data= rads
        self.publisher_counter_.publish(new_msg)
        self.get_logger().info(f"Recibido:{rpm:.2f} RPM -» Convertido:{rads:.2f} rad/s")

def main(args= None):
    rclpy.init(args=args)
    node = NodeCounter()
    rclpy.spin(node)
    rclpy.shutdown()

if _name_ == 'main':
    main()
```
