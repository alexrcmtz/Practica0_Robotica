## Previo de la práctica

- #### **URDF** [1], [2]

Un **URDF (Unified Robot Description Format)** es un archivo XML que describe la estructura física y cinemática de un robot. Lo podemos ver como un "plano" digital del robot. Asu vez, un **XML (eXtensible Markup Language, Lenguaje de Marcado Extensible)** es un formato de texto para almacenar y organizar datos de forma estructurada y legible.

Según ROS2, un **URDF** especifíca:

- Geometría del robot, la forma de cada eslabón del que se conforma.
- Estructura del robot y cómo se conectan cada parte.
- Propiedades físicas tales como la inercia, masa o fricción.
- Sensores y actuadores, así como donde se posicionan y cómo fucnionan.
- Y apariencias visuales como colores y texturas.

Se conforman de tres elementos clave:

1. **Links (eslabones o vínculos)**
   Éstos representan las partes rígidas del robot y cada link se caracteriza de lo siguiente:
   
   - Nombre único.
   - Forma visual (como box, cylinder, sphere o mesh).
   - Propiedades físicas como masa o inercia.
   - Texturas y colores.
    
    Un ejemplo se muestra a continuación:

    ```xml
    <link name="base_link">
        <inertial>
            <mass value="2.0"/>
            <inertia ixx="0.01" ixy="0" ixz="0" 
                    iyy="0.01" iyz="0" izz="0.01"/>
        </inertial>
        <visual>
            <geometry>
                <box size="0.2 0.2 0.1"/>
            </geometry>
            <material name="grey">
                <color rgba="0.5 0.5 0.5 1.0"/>
            </material>
        </visual>
    </link>
    ```

- `<link>` define una parte rígida del robot
- `<inertial>` especifica las propiedades físicas de masa e inercia
- `<visual>` define cómo se ve el link

2. **Joints o articulaciones**
   Representan las conexiones entre links y definen cómo se mueven las partes, los tipos de juntas son:

   - Revolute. Que son la rotación o giros de motores típicos.
   - Prismatic. Tienen movimniento lineal o traslacional, como un pistón.
   - Fixed. No tiene movimiento pues es fijo.
   - Continuos. Tiene rotación continua ilimitada.
   - Planar. Tienen movimiento en dos dimensiones.
    
    Un ejemplo es el siguiente:

    ```xml
    <joint name="shoulder_joint" type="revolute">
        <parent link="base_link"/>
        <child link="upper_arm"/>
        <origin xyz="0 0 0.1" rpy="0 0 0"/>
        <axis xyz="0 1 0"/>
        <limit lower="-1.57" upper="1.57" effort="10" velocity="1"/>
    </joint>
    ```

3. **Transformaciones (frames o coordenadas)**
   Definen la psoción y orientación relativa entre links, usan lo siguiente:

   - `xyz = x, y, z` (distancia en metros)
   - `rpy = roll, pitch, yaw` (ángulos en radianes)

La estructura general de un URDF se puede mostrar a continuación:

```xml
<?xml version="1.0"?>
<robot name="robot">
    
    <!-- LINKS -->
    <link name="base_link">
        <!-- Propiedades visuales y físicas -->
    </link>
    
    <link name="wheel_left">
        <!-- Propiedades visuales y físicas -->
    </link>

    <link name="wheel_right">
        <!-- Propiedades visuales y físicas -->
    </link>
    
    <!-- JOINTS -->
    <joint name="wheel_right_joint" type="continuous">
        <parent link="base_link"/>
        <child link="wheel_left"/>
        <origin xyz="0 0.15 0.05" rpy="0 0 0"/>
        <axis xyz="0 1 0"/>
    </joint>

    <joint name="wheel_left_joint" type="continuous">
        <parent link="base_link"/>
        <child link="wheel_left"/>
        <origin xyz="0 0.15 0.05" rpy="0 0 0"/>
        <axis xyz="0 1 0"/>
    </joint>
    
</robot>
```

A continuación se muestra el URDF de un robot móvil simple conformado por la caja del robot y dos ruedas:

```xml
<?xml version="1.0"?>
<robot name="robot_movil">
    
    <!-- Cuerpo principal -->
    <link name="base_link">
        <visual>
            <geometry>
                <box size="0.3 0.2 0.1"/>
            </geometry>
            <material name="blue">
                <color rgba="0 0 1 1"/>
            </material>
        </visual>
    </link>
    
    <!-- Rueda izquierda -->
    <link name="wheel_left">
        <visual>
            <geometry>
                <cylinder length="0.05" radius="0.05"/>
            </geometry>
            <material name="black">
                <color rgba="0 0 0 1"/>
            </material>
        </visual>
    </link>
    
    <!-- Rueda derecha -->
    <link name="wheel_right">
        <visual>
            <geometry>
                <cylinder length="0.05" radius="0.05"/>
            </geometry>
            <material name="black">
                <color rgba="0 0 0 1"/>
            </material>
        </visual>
    </link>
    
    <!-- Conección de la rueda izquierda al cuerpo -->
    <joint name="wheel_left_joint" type="continuous">
        <parent link="base_link"/>
        <child link="wheel_left"/>
        <origin xyz="0 0.1 0.05" rpy="1.57 0 0"/>
        <axis xyz="0 0 1"/>
    </joint>
    
    <!-- Conección de la rueda derecha al cuerpo -->
    <joint name="wheel_right_joint" type="continuous">
        <parent link="base_link"/>
        <child link="wheel_right"/>
        <origin xyz="0 -0.1 0.05" rpy="1.57 0 0"/>
        <axis xyz="0 0 1"/>
    </joint>
    
</robot>
```

En el ejemplo anterior se ve la descripción del robot llamado robot_móvil, se contemplan los eslabones de los que se conforma (y las características visuales de cada uno), así como de las juntas y sus transformaciones.

Usos del **URDF**

1. Simulación en Gazebo
   URDF -> Gazebo -> Simula al robot con física realista.

2. Visualización en RViz
   URDF -> RViz -> Visualiza el robot en 3D.

3. Control y cinemática
   El URDF define la estructura y ROS2 calcula la cinemática directa e inversa.

4. Planificación de movimiento
   El URDF describe las juntas y el planificador mueve al robot.

5. Sensores virtuales
   Se pueden agregar cámaras , lidar, IMUS o más al URDF.



- #### **RViz** [3]

**RViz o ROS Visualization** es una herramienta de visualización 3D que nos permite ver y entender que hace un robot en tiempo real. RViz nos permite:

- Ver el modelo del robot basado en el URDF.
- Visualizar datos de sensores como cámaras, ultrasónicos, etc.
- Ver el sistema de coordenadas (frames).
- Analizar trayectorias de movimiento.
- Debuggear problemas del robot.
- Monitorear datos en tiempo real.
- Visualizar velocidades, aceleraciones y fuerzas.

- #### **XACRO** [4], [5]

Un archivo **XACRO ó XML Macros** es una extensión de URDF que añade programación, como variables, loops o condiciones, a los archivos XML, permitiendo hacer robots más complejos de forma más simple. Sus características principales se enlistan a continuación:

- **Variables** `<xacro: property>`. Se definen variables que se pueden usar en todo el archivo. Por ejemplo:

    ```xml
    <xacro:property name="pi" value="3.14159"/>
    <xacro:property name="robot_mass" value="10.0"/>

    <!-- Usar la variable -->
    <mass value="${robot_mass}"/>
    ```

- **Macros** `<xacro: macro>`. Se definen funciones reutilizables. Por ejemplo:
    ```xml
    <xacro:macro name="crear_cilindro" params="nombre largo radio">
        <link name="${nombre}">
            <visual>
            <geometry>
                <cylinder length="${largo}" radius="${radio}"/>
            </geometry>
            </visual>
        </link>
    </xacro:macro>

    <!-- Usar la macro -->
    <xacro:crear_cilindro nombre="eslabón_1" largo="0.5" radio="0.1"/>
    <xacro:crear_cilindro nombre="eslabón_2" largo="0.3" radio="0.08"/>
    ```

- **Includes** `<xacro: include>`. Importar otros archivos XACRO, por ejemplo:
    ```xml
    <!-- materials.xacro -->
    <xacro:property name="color_rojo">
        <color rgba="1 0 0 1"/>
    </xacro:property>

    <!-- robot.xacro -->
    <xacro:include filename="materials.xacro"/>
    <link name="base">
        <visual>
            <material name="rojo"/>
        </visual>
    </link>
    ```
- **Loops** `<xacro: if>` y `xacro: foreach`. Crear objetos repetidos, por ejemplo:
    ```xml
    <!-- Crear 5 eslabones -->
    <xacro:foreach params="i" values="1 2 3 4 5">
        <xacro:crear_cilindro nombre="link_${i}" largo="0.5" radio="0.1"/>
    </xacro:foreach>
    ```

- **Expresiones matemáticas** `${...}`. Se pueden realizar cálculos, por ejemplo:
    ```xml
    <xacro:property name="ancho" value="0.1"/>
    <xacro:property name="largo" value="${ancho * 5}"/>  <!-- 0.5 -->

    <origin xyz="0 0 ${ancho/2}"/>  <!-- 0.05 -->
    ```

- #### **YAML** [6], [7]

Un archivo **YAML (YAML Ain´t MArkup Language)** es un formato de configuración legible para ususarios mucho más sencillo que JSON´s o XML's. Tienen la siguiente estructura básica:

- **Pares clave: valor**
    ```yaml
    nombre: Poncho
    edad: 25
    ciudad: México
    ```

- **Listas con -**
    ```yaml
    robots:
      - nombre: robot_1
        eslabones: 3
      - nombre: robot_2
        eslabones: 5
    ```
    ó en una sola línea

    ```yaml
    colores: [rojo, verde, azul]
    ```

- **Anidación con indentación**
    ```yaml
    robot:
    base:
    masa: 5.0
    radio: 0.1
    eslabones:
      - nombre: link_1
        largo: 0.5
      - nombre: link_2
        largo: 0.3
    ```

- **Tipos de datos**
    ```yaml
    cadena: "Hola ROS2"
    numero: 42
    decimal: 3.14159
    booleano: true
    nulo: null
    lista: [1, 2, 3, 4, 5]
    ```

Los YAML se usan para configurar parámetros de nodos, controladores, y sistemas. En lugar de escribir código, se define la configuración. Se muestra un ejemplo de parámetros de un nodo y de una configuración de sensores:

```yaml
# controller.yaml
robot_controller:
  ros__parameters:
    joint_names: ["joint_1", "joint_2", "joint_3"]
    update_rate: 50.0  # Hz
    max_velocity: [1.57, 1.57, 1.57]  # rad/s
    max_effort: [10.0, 10.0, 10.0]    # N


# sensor_config.yaml
lidar:
  frame_id: "base_link"
  scan_topic: "/scan"
  range:
    min: 0.1
    max: 30.0
  frequency: 10  # Hz

camera:
  frame_id: "camera_link"
  image_topic: "/camera/image"
  resolution:
    width: 640
    height: 480
```

- #### **Gazebo** [8], [9]

**Gazebo** es un simulador de físicas 3D para robots; realiza tres pincipales acciones:

- **Simulación de físicas**
    Calcula movimientos realistas según las lesyes de la física.
    - Gravedad.
    - Inercia.
    - Fricciones.
    - Colisiones.
    - Fuerzas.
    - Torques.

- **Simulación de sensores**
    Genera datos como si fuesen reales.
    - Cámara: genera imágenes RGB.
    - LIDAR: genera puntos de nubes 3D.
    - IMU: genera aceleraciones y rotaciones.
    - Micrófono: genera audio.

- **Visualización 3D**
    Muestra el mundo en forma visual.
    - El propio robot.
    - El entorno que rodea al robot.
    - Luces y sombras.

**Gazebo** usa dos tipos de archivos:

  - Archivos `.world`: se definen como mundos o escenarios que definen el entorno completo. Contiene configuraciones físicas, iluminación, objetos estáticos y dinámicos.
  - Archivos `.sdf` o extensiones XACRO: definen propiedades de simulación específicas del robot.

  - **`gz2_scara.launch.py`** es el archivo que coordina toda la ejecución del robot, es el "orquestador de todo y podemos verlo en la siguiente parte:

    ```python
    def generate_launch_description():
        # Paso 1: Obtiene rutas de ambos paquetes
        pkg_scara_description = get_package_share_directory('scara_description')
        pkg_scara_bringup = get_package_share_directory('scara_bringup')
    
        # Paso 2: Define rutas específicas
        urdf_path = os.path.join(pkg_scara_description, 'urdf', 'gz2_scara.xacro')
        gazebo_config_path = os.path.join(pkg_scara_bringup, 'config', 'gz_bridge.yaml')
        world_path = os.path.join(pkg_scara_bringup, 'worlds', 'world_config.world')
    
        # Paso 3: Lanza 5 nodos en orden
        return LaunchDescription([
            robot_state_publisher_node,   # 1. Publica el URDF
            gz_sim,                       # 2. Abre Gazebo
            spaw_entity,                  # 3. Crea el robot en Gazebo
            gz_ros_bridge_node,           # 4. Puente Gazebo ↔ ROS2
            rviz_node                     # 5. Abre RViz
    ])
    ```

  - **`gz_bridge.yaml`** funciona como un puente de comunicación entre Gazebo y ROS2, tiene la función de traductor.

    ```yaml
    # SENSORES: Gazebo → ROS2 (lo que el robot "ve")

      gz_topic_name: "/world/world_config/model/robot_scara/joint_state"

      gz_topic_name: "/camera/image_raw"
      direction: GZ_TO_ROS

    - ros_topic_name: "/scan"
      gz_topic_name: "/scan"
      direction: GZ_TO_ROS

    # COMANDOS: ROS2 → Gazebo (órdenes que el robot recibe)
    - ros_topic_name: "/joint1/cmd_pos"
      gz_topic_name: "/model/robot_scara/joint/link_1_joint/0/cmd_pos"
      direction: ROS_TO_GZ  # De ROS2 a Gazebo

    - ros_topic_name: "/joint2/cmd_pos"
      gz_topic_name: "/model/robot_scara/joint/link_2_joint/0/cmd_pos"
      direction: ROS_TO_GZ

    - ros_topic_name: "/joint3/cmd_pos"
      gz_topic_name: "/model/robot_scara/joint/link_3_joint/0/cmd_pos"
      direction: ROS_TO_GZ
    ```

A través de la estructuración sistemática de paquetes y directorios clave, se validó la importancia de la organización jerárquica para garantizar la portabilidad y el correcto funcionamiento del sistema.

Durante el análisis arquitectónico, se profundizó en la descripción cinemática y dinámica del robot mediante archivos URDF y su optimización a través de macros XACRO. El uso de XACRO y archivos de configuración YAML demostró ser una estrategia de ingeniería crítica para parametrizar el modelo, reducir la redundancia de código y facilitar la integración de elementos complejos como articulaciones y sensores.

Asimismo, se evaluó el mecanismo de despliegue automatizado mediante scripts de lanzamiento (Launch files) en Python. Desde una perspectiva sistémica, el correcto diseño del launch abstrae la complejidad de la inicialización, coordinando en un solo comando la carga del modelo, la sincronización temporizada y el arranque simultáneo del entorno gráfico y los nodos de control, eliminando la necesidad de configuraciones manuales posteriores en los simuladores.

En cuanto a la simulación física, se determinó el rol fundamental del entorno Gazebo y el puente de comunicación (ros_gz_bridge). Se comprendió cómo la correcta declaración de propiedades de masa, inercia y controladores de posición (PID) dentro del entorno virtual permite aproximar el comportamiento dinámico del robot a las restricciones del mundo real.

Finalmente, el aprendizaje adquirido proporciona un panorama metodológico sólido para el flujo de trabajo CAD-to-ROS 2. Se identificó el potencial de diseñar manipuladores y sistemas móviles en software de diseño asistido por computadora para su posterior exportación a formatos URDF/XACRO. Esta integración entre el diseño mecánico y la programación en ROS 2 es fundamental para el prototipado virtual y el análisis de geometrías complejas en proyectos robóticos de nivel profesional.
